```python
import pandas as pd
dd=pd.read_csv("training_diabetes.csv")
dd.head()
df=dd.copy()
df.head()
df.tail()
```

```python
x=df.iloc[0:331,0:10]

y=df.Y
y
x
```

```python
x
```

```python
from sklearn import model_selection
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
alg=LinearRegression()
alg.fit(x_train,y_train)
```

```python
y_pre=alg.predict(x_test)
```

```python
print(alg.score(x_train,y_train))
print(alg.score(x_test,y_test))
```

```python
import numpy as np
daa=np.genfromtxt("test.csv",delimiter=",")
test_pre=alg.predict(daa)
print(test_pre)
```
