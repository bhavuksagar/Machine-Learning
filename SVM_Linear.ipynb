```python
import numpy as np
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

```python
x= np.array([[1,1],[1,2],[2,1],[1.5,1.5],[6,4],[5,6],[5,5],[6,5],[7,6],[7,7]])
y=[0,0,1,0,1,1,1,1,1,1]
```

```python
x_x1=x[:,0]
x_x2=x[:,1]
plt.scatter(x_x1,x_x2,c=y)
plt.show()
```

```python
clg=SVC(kernel='linear',C=10000).fit(x,y)
clg.coef_,clg.intercept_
```

```python
x1=np.array([0,5])
x2=-1*(clg.intercept_ + clg.coef_[0][0] *x1)/clg.coef_[0][1]
plt.plot(x1,x2,'r')
plt.scatter(x_x1,x_x2,c=y)
plt.axis([0,8,0,8])
plt.show()
```
