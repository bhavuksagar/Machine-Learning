```python
from sklearn import datasets
```

```python
#loading the data

data = datasets.load_breast_cancer()
```

```python
data
```

```python
x=data.data
y=data.target
```

```python
x,y
```

```python
from sklearn import model_selection
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
clg=LogisticRegression()
```

```python
#training the module
clg.fit(x_train,y_train)
```

```python
y_pre=clg.predict(x_test)
y
```

```python
#checking the accuracy
print("Testing data Score:",clg.score(x_test,y_test))
print("Training data Score:",clg.score(x_train,y_train))
```

```python
#checking that where we are doing mistakes. The non zero values are the wrong predict values
y_test - y_pre
```

```python
#checking the probability of predicting the values at each point

train_prob=clg.predict_proba(x_train)
test_prob=clg.predict_proba(x_test)

train_prob
```
