```python
from nltk.corpus import movie_reviews
```

```python
document=[]
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        document.append((movie_reviews.words(fileid),category))
document[0:5]
        
```

```python
import random
random.shuffle(document)
document[0:5]
```

```python
from nltk.corpus import stopwords
import string
stop=set(stopwords.words("english"))
pun=list(string.punctuation)
stop.update(pun)
stop
```

```python
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def clean(word):
    clean_words=[]
    for w in word:
        if w not in stop:
            pos=pos_tag([w])
            output=lemmatizer.lemmatize(w,pos=get_simple(w))
            clean_words.append(output.lower())
    return clean_words
```

```python
document=[(clean(doc),category)for doc,category in document]
```

```python
document[0:5]
```

```python
category=[category for document,category in document]
```

```python
text=[" ".join(doc) for doc,category in document]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(text,category)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2000,ngram_range=(1,2))
x_train_feature=cv.fit_transform(x_train)
```

```python
x_train_feature.todense()
```

```python
cv.get_feature_names()
```

```python
x_test_feature=cv.transform(x_test)
```

```python
from sklearn.svm import SVC
clg=SVC(kernel='linear')
clg.fit(x_train_feature,y_train)
```

```python
clg.score(x_test_feature,y_test)
```
